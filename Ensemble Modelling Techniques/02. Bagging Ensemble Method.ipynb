    Bagging, short for Bootstrap Aggregating, is a popular ensemble learning method designed to improve the accuracy and robustness of machine learning models.
    
    It works by creating multiple instances of a base learning algorithm, training each on different subsets of the training data, and then combining their predictions through an aggregation process.

    Base Learner are decision trees.
    
```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

# Create a base estimator (in this case, a decision tree)
base_estimator = DecisionTreeClassifier()

# Create a BaggingClassifier using the base estimator
bagging_classifier = BaggingClassifier(base_estimator, n_estimators=10, random_state=42)

# Train the BaggingClassifier on your training data
bagging_classifier.fit(X_train, y_train)

# Make predictions using the trained BaggingClassifier
predictions = bagging_classifier.predict(X_test)
```    

    A Bagging classifier is an ensemble meta-estimator that fits base classifiers each on random subsets of the original dataset and then aggregate their individual predictions (either by voting or by averaging) to form a final prediction.

    Bagging classifier is nothing but bootstrap aggregation.

    Bootstrap Sampling :
        It involves creating multiple subsets of the training data by randomly sampling with replacement. 
        
    Base Learner Construction: 
        An algorithm, such as decision trees or other models, is trained independently on each bootstrap sample. This results in multiple diverse base learners.
    
    Predictions Combination: 
        Once the individual base learners are trained, Bagging combines their predictions through aggregation. 
        This aggregation is Voting method where we choose major number outcome from model.                

    Advantages of Bagging:

    Increased accuracy: 
        By combining predictions from multiple models, Bagging often improves predictive performance compared to individual models.
    Robustness: 
        Bagging helps mitigate overfitting and reduces the impact of noisy data points or outliers.

    Limitations:

    Computational complexity: 
        Training multiple models can be computationally expensive, especially for large datasets.
    Interpretability: 
        The combined model might be less interpretable compared to individual models.

# Random Forest
    One of the most well-known applications of Bagging is the Random Forest algorithm, which employs a collection of decision trees trained via Bagging. 
    Each tree is grown on a bootstrapped dataset, and during tree construction, a random subset of features is considered at each split to further enhance diversity among the trees.

    The Forest in Random Forest is densely populated data.

    If 
        m-> number of subsets of predictor
        p-> number of predictor
    then
        m = p -> Bagging Classifer
        m < p -> Random Forest
