### relplot (Relational Plot):

    Type: 
        Plots for exploring relationships between two numeric variables.

    Description: 
        relplot is a versatile function in Seaborn that allows you to create scatter plots or line plots to explore the relationship between two numeric variables. 
        It is a high-level interface that combines FacetGrid with plotting functions.

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_length_mm")
plt.title("replot")
plt.show()
```

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue='sex')
plt.title("Relationship Between Flipper Length and Bill Length in Penguins (Colored by Sex)")
plt.show()
```

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue='sex', col='sex')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Create a line plot using Seaborn's lineplot
sns.lineplot(data=flights_data, x="year", y="passengers", hue="month")

# Set labels and title
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.title("Passenger Numbers Over Time by Month")

# Display the plot
plt.show()
```

### catplot (Categorical Plot):

    Type: 
        Categorical plots for showing the relationship between a numeric variable and one or more categorical variables.
    
    Description: 
        catplot is used to create different types of categorical plots such as scatter plots, bar plots, box plots, etc., by specifying the kind parameter.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Create a scatter plot using Seaborn's scatterplot
sns.scatterplot(data=flights_data, x="year", y="passengers", hue="month", s=100, alpha=0.8)

# Set labels and title
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.title("Scatter Plot of Passenger Numbers Over Time by Month")

# Display the legend
plt.legend(title="Month", loc="upper left", bbox_to_anchor=(1, 1))

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "penguins" dataset
penguins = sns.load_dataset("penguins")

# Create a strip plot using Seaborn's stripplot
sns.stripplot(data=penguins, x="species", y="bill_length_mm", hue="sex", jitter=True, dodge=True, alpha=0.7)

# Set labels and title
plt.xlabel("Species")
plt.ylabel("Bill Length (mm)")
plt.title("Strip Plot of Bill Length by Species and Sex")

# Display the legend
plt.legend(title="Sex", loc="upper right", bbox_to_anchor=(1.25, 1))

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "penguins" dataset
penguins = sns.load_dataset("penguins")

# Create a swarm plot using Seaborn's swarmplot
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue="sex", dodge=True, palette="Set2")

# Set labels and title
plt.xlabel("Species")
plt.ylabel("Bill Length (mm)")
plt.title("Swarm Plot of Bill Length by Species and Sex")

# Display the legend
plt.legend(title="Sex", loc="upper right", bbox_to_anchor=(1.25, 1))

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "penguins" dataset
penguins = sns.load_dataset("penguins")

# Create a violin plot using Seaborn's violinplot
sns.violinplot(data=penguins, x="species", y="bill_length_mm", palette="Set2", split=True)

# Set labels and title
plt.xlabel("Species")
plt.ylabel("Bill Length (mm)")
plt.title("Violin Plot of Bill Length by Species")

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create a point plot using Seaborn's pointplot
sns.pointplot(data=diamonds, x="cut", y="price", hue="color", markers="s", linestyles="-", dodge=True)

# Set labels and title
plt.xlabel("Cut")
plt.ylabel("Price")
plt.title("Point Plot of Diamond Price by Cut and Color")

# Display the legend
plt.legend(title="Color", loc="upper right", bbox_to_anchor=(1.25, 1))

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset
tips = sns.load_dataset("tips")

# Create a bar plot using Seaborn's barplot
sns.barplot(data=tips, x="day", y="total_bill", ci=None, palette="Set3")

# Set labels and title
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.title("Average Total Bill by Day of the Week")

# Display the plot
plt.show()
```

### displot (Distribution Plot):

    Type: 
        Visualizing univariate distributions of a single variable or comparing distributions between two or more variables.
    
    Description: 
        displot is a figure-level function for creating various univariate distribution plots such as histograms, kernel density plots, and cumulative distribution functions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset
tips = sns.load_dataset("tips")

# Example: Histogram using displot
tips = sns.load_dataset("tips")
sns.displot(tips["total_bill"], bins=20, kde=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram using Seaborn's histplot
plt.figure(figsize=(8, 6))
sns.histplot(diamonds["price"], bins=30, kde=True, color="skyblue")
plt.title("Histogram of Diamond Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create a distribution plot using Seaborn's displot
plt.figure(figsize=(8, 6))
sns.displot(diamonds["price"], kde=True, color="salmon", bins=30)
plt.title("Distribution Plot of Diamond Prices")
plt.xlabel("Price")
plt.ylabel("Density")
plt.show()
```
