    Here we use matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

tips = sns.load_dataset('tips')
```

```python
tips.info()
```

# Univariate
    “univariate” refers to a single variable.

```python
plt.hist(tips['total_bill'], bins= 10)
plt.title('Histogram of Total Bill')
plt.show()
```

# Bivariate
    “bivariate” refers to a two variables.
    
## Scatter plot    

```python
plt.scatter(data=tips, x='total_bill', y='tip')
plt.title('Scatterplot of Total Bill vs Tip')
plt.show()
```

## Boxplot
    Boxplots are used when a discrete variable is plotted against a continuous variable

```python
tips.boxplot(column='tip', by='sex', grid=False)
plt.title('Boxplot of Total Bill by Day')
plt.show()
```
