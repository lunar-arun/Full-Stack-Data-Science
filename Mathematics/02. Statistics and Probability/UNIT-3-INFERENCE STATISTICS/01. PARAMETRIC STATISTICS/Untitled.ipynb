### Parametric statistics 

    Refers to a branch of statistics that assumes the underlying data follows a specific probability distribution.
    
    In other words, it involves making assumptions about the population distribution from which the sample is drawn.
    
    These methods often assume that the data is normally distributed, and they rely on the estimation of parameters such as mean and standard deviation.

#### ESTIMATING THE POPULATION MEAN USING THE z STATISTIC ( KNOWN)

    A point estimate is a statistic taken from a sample that is used to estimate a population parameter.
    
    An interval estimate (confidence interval) is a range of values within which the analyst can declare, with some confidence, the population parameter lies.
    
$$z = \frac{{x - {m}}}{{s/\sqrt{n}}}$$
