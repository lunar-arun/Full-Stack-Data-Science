    Statistics : a science dealing with the collection, analysis, interpretation, and presentation of numerical data.
    
    Two Branch :
        Descriptive Statistics        Inferential Statistics
        
    Before we go in, lets understand population and sample.
    
    Population : a collection of persons, objects, items of interest.
    
    Sample : a portion of the whole from items of interest.
    
    Census : gather data from the whole population for a given measurement of interest.
    
    Descriptive Statistics : using data gathered on a group to describe or -reach conclusions- about that same group.
        
    Inferential Statistics : gathers data from a sample and uses the statistics generated to reach conclusions about the population from which the sample was taken.
    
    The advantage of using inferential statistics is that they enable the researcher to study effectively a wide range of phenomena without having to conduct a census

    In inferential statistics :
        a descriptive measure of population is called : Parameter
        
        a descriptive measure of sample is called : Statistic
        
    its only used in inferential statistics.

    Data Measurement : There are four common level of measurement.
    
    1. Nominal Level 
        Used only to classify without any particular order or ranking.
        Deals with qualitative data.
        Such as gender, eye color.
        
    2. Ordinal Level
        Used to classify with meaningful order or ranking.
        Such as ratings, or levels of agreement(e.g., low, medium, high)
        
    3. Interval Level
        Used when data that has a meaningful order, consistent intervals between values, but lacks a true zero point.
        Such temperature.
        
    4. Ratio Level
        It possesses all the characteristics of the interval level but includes a true zero point.
        Such as weight, height, time.

    Charts and Graphs
    
    To build this we need raw data and they are in two forms.
    
    Ungrouped Data :
        Raw data, or data that have not been summarized in any way.
        
    Grouped Data :
        Data that have been organized into a frequency distribution.
    
    
    Frequency Distribution :
        a summary of data presented in the form of class intervals and frequencies.
        
        Class intervals :
            ranges that divide the data into groups or classes.
            
        Frequencies : 
            number of times a set of values occurs within each class interval.
        
        Range : difference between largest and smallest numbers.

    Class midpoint : midpoint of each class interval.

    Relative frequency : proportion of the total frequency that is in any given class interval in a frequency distribution.

    Cumulative frequency : running total of frequencies through the classes of a frequency distribution.

```python
import pandas as pd
import numpy as np

# Ungrouped data (Unemployment scores)
data = [2.3, 7.0, 6.3, 11.3, 9.6, 2.8, 7.1, 5.6, 10.6, 9.1, 3.6, 5.9, 5.4, 9.7, 8.3, 2.4, 5.5, 7.1, 8.8, 7.6, 2.9, 4.7, 7.1, 7.8, 6.8, 3.0, 3.9, 8.0, 7.5, 7.2, 4.6, 3.6, 8.4, 8.1, 7.7, 4.4, 4.1, 7.5, 10.3, 7.6, 3.4, 4.8, 7.5, 11.2, 7.2, 4.6, 4.7, 7.6, 11.4, 6.8, 6.9, 5.9, 11.0, 10.4, 6.3, 6.0, 6.4, 12.0, 9.5, 6.0]

# Determine the range of the data
data_range = np.ptp(data) # np.ptp(peak to peak) nothing but "range"

# Decide on the number of bins (intervals) for the grouped data
num_bins = 6

# Calculate the bin width
bin_width = data_range / num_bins

# Define the bin edges
bin_edges = [min(data) + i * bin_width for i in range(num_bins + 1)]

# Create a histogram to get the frequency distribution
hist, _ = np.histogram(data, bins=bin_edges)

# Create a DataFrame to display the grouped data
grouped_data = pd.DataFrame({
    'Bin Edges': [f'{bin_edges[i]} - {bin_edges[i+1]}' for i in range(num_bins)],
    'Frequency': hist
})

grouped_data
```

#### QUALITATIVE DATA GRAPHS

    Pie Chart : 
        A circular depiction of data where the area of the whole pie represents 100% of the data and slices of the pie represent a percentage breakdown of the sublevels.

```python
import matplotlib.pyplot as plt

# Sample data for the pie chart
sizes = [30, 20, 15, 35]  # Portions for each slice of the pie
labels = ['A', 'B', 'C', 'D']  # Labels for each slice

plt.figure(figsize=(4, 4))  # Set the figure size (optional)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green', 'orange'])
plt.title('Sample Pie Chart')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

     Bar Graphs : 
         A bar graph or chart contains two or more categories along one axis and a series of bars, one for each category, along the other axis.

```python
import matplotlib.pyplot as plt

# Sample data for the bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 35]

plt.figure(figsize=(8, 6))  # Set the figure size (optional)
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Sample Bar Chart')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed

plt.show()
```
