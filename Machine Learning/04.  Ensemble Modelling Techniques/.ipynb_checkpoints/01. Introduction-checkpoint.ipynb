    Ensemble learning is a technique that combines multiple machine learning algorithms to produce one optimal predictive model.
        TYPES OF ENSEMBLE MODELING TECHNIQUES
            Bagging (with reduced variance)
                 Uses samples to get an estimate by aggregating the scores over many samples. 
                 It's effective when you have limited data.
            Boosting (with reduced bias)
                Boosting involves sequentially adding models to the ensemble where 
                subsequent models correct the performance of prior models.
            Stacking (with improved predictions)
                Combines multiple machine learning algorithms via meta learning. 
            Blending
                technique that uses a machine learning model to learn how to best combine the predictions from multiple contributing ensemble member models.

    Ensemble learning techniques are utilized to improve the performance of machine learning models by combining multiple models together. 
    These methods help in overcoming various limitations that individual models might face, such as:
    
        Bias and Variance Trade-off:
            Individual models suffer from either high bias(underfitting) or high variance(overfitting).
            
            Ensemble methods aim to strike a balance between bias and variance, resulting in improved generalization gap and performance.
            
        Overfitting: 
            Ensemble methods reduce overfitting by aggregating predictions from multiple models, thereby reducing the chance of learning noise present in the training data.
            
        Handling Imbalanced Data:
            Ensemble methods can handles better than individual models by giving more weight or importance to minority class samples, hence improving classification performance.
